```python
import pandas as pd
import numpy as np
from sodapy import Socrata
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
client = Socrata("www.datos.gov.co", None)
sincelejo=pd.DataFrame(client.get_all("gt2j-8ykr",content_type='json?ciudad_municipio_nom=SINCELEJO'))
barranquilla=pd.DataFrame(client.get_all("gt2j-8ykr",content_type='json?ciudad_municipio_nom=BARRANQUILLA'))
```

```python
sincelejo['fecha_reporte_web']=pd.to_datetime(sincelejo['fecha_reporte_web'],format="%d/%m/%Y %H:%M:%S")
barranquilla['fecha_reporte_web']=pd.to_datetime(barranquilla['fecha_reporte_web'],format="%d/%m/%Y %H:%M:%S")
casosincelejo=sincelejo['fecha_reporte_web'].value_counts(sort=False).to_frame('Casos Positivos Sincelejo')
casosbarranquilla=barranquilla['fecha_reporte_web'].value_counts(sort=False).to_frame('Casos Positivos Barranquilla')
```

```python
fig,(ax1,ax2)=plt.subplots(2,figsize=(20,10))
ax1.bar(casosincelejo.index,casosincelejo['Casos Positivos Sincelejo'],width=1)
ax1.set_title(' Casos historicos Covid Sincelejo')
ax1.set_xlabel('Fecha de reporte')
ax1.set_ylabel('# de casos')
ax2.bar(casosbarranquilla.index,casosbarranquilla['Casos Positivos Barranquilla'],width=1)
ax2.set_xlabel('Fecha de reporte')
ax2.set_ylabel('# de casos')
ax2.set_title(' Casos historicos Covid Barranquilla')
plt.show()
```

```python
GeneroSincelejo=sincelejo['sexo'].value_counts()
GeneroBarranquilla=barranquilla['sexo'].value_counts()
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(10,4))
ax1.set_title('Casos positivos Covid Sincelejo',fontsize=13)
ax1.pie(GeneroSincelejo,labels=GeneroSincelejo.index,colors=['#F50CE3','#0D52FF'],explode=[0,0.1],autopct='%1.1f%%')
ax1.text(-0.30,0.1,str(GeneroSincelejo[0]),fontsize=20)
ax1.text(-0.30,-0.4,str(GeneroSincelejo[1]),fontsize=20)
ax2.set_title('Casos positivos Covid Barranquilla',fontsize=13)
ax2.pie(GeneroBarranquilla,labels=GeneroBarranquilla.index,colors=['#F50CE3','#0D52FF'],explode=[0,0.1],autopct='%1.1f%%')
ax2.text(-0.30,0.1,str(GeneroBarranquilla[0]),fontsize=20)
ax2.text(-0.30,-0.4,str(GeneroBarranquilla[1]),fontsize=20)
plt.show()
```

# Anlisis
## Se ha demostrado entonces como la proporcin de mujeres en cada ciudad correspondientemente es la que muestra un mayor numero de casos con respecto a los hombres.

<hr>

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,5))
datos1=np.array(sincelejo['edad'].astype(int))
dicc=ax1.boxplot(datos1,showmeans=True)

m1=datos1.mean(axis=0)
st1=datos1.std(axis=0)
for i, line in enumerate(dicc['medians']):
    x, y = line.get_xydata()[1]
    ax1.text = ' ={:.2f}\n ={:.2f}'.format(m1, st1)
    ax1.annotate(ax1.text, xy=(x, y))
    
datos2=np.array(barranquilla['edad'].astype(int))
dicc2=ax2.boxplot(datos2,showmeans=True)
m2=datos2.mean(axis=0)
st2=datos2.std(axis=0)
for i, line in enumerate(dicc2['medians']):
    x, y = line.get_xydata()[1]
    ax2.text = ' ={:.2f}\n ={:.2f}'.format(m2, st2)
    ax2.annotate(ax2.text, xy=(x, y))
```
