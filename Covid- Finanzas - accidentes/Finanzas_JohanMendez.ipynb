# Comparación de las acciones
## La gráfica muestra con claridad como Tesla es la poosedora de más acciones entre todas las compañias en el presente, ademas de ser la mas fluctuante dejando en evidencia a simple vista su alta desviación estandar.

## De igual manera podemos apreciar que las acciones de EC(Ecopetrol) son las menores de todas.
<hr>

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

cerrados=data['Close']
hoy=datetime.datetime.today()
informacion=cerrados.to_numpy()

Tabla50=pd.DataFrame({})
Tabla100=pd.DataFrame({})
#Graficando
fig,axes=plt.subplots(informacion.shape[1],figsize=(20,20))
for i in range(informacion.shape[1]):
    axes[i].plot(cerrados.iloc[:,[i]])
    axes[i].set_title('Graficas acciones de '+cerrados.columns[i])
    dias50=np.flipud(moving_average(informacion[:,i],50))
    Tabla50[cerrados.columns[i]]=pd.DataFrame(dias50)
    dias100=np.flipud(moving_average(informacion[:,i],100))
    Tabla100[cerrados.columns[i]]=pd.DataFrame(dias100)
    axes[i].plot([hoy -datetime.timedelta(days=x*1.45) for x in range(dias50.shape[0])],dias50)
    axes[i].plot([hoy - datetime.timedelta(days=x*1.45) for x in range(dias100.shape[0])],dias100)
    axes[i].legend(['Acciones cerradas '+cerrados.columns[i],'Media móvil 50 días','Media móvil 100 días'])
    axes[i].set_ylabel('Acciones')
plt.show()    
```

# Analisis de la gráfica
## La linea azul representa el cierre de las acciones en el tiempo, a partir de estos datos diarios se saca la media movil de 50 días y 100 días la cual sera un indicador de tendencia a lo largo del grafico permitiendonos visualizar sus movimientos al suavizar la gráfica, esto ayudará a decidir si invertir a pesar de tener un desviación estandar tan alta, pues si las acciones de tesla poseen esta variable tan elevada la pendiente de sus medias moviles indican el alza, **lo que hace a TSLA la mejor opción** (23/08/2021) por poseer los cierres de acciones más altos con una desviación estandar a favor por la media móvil con promedios mayores a la media de la información original.
